```python
##most of the code for reading the radar files was borrowed from examples here
##https://github.com/Unidata/pyudl/tree/master/examples/notebooks/radar

##shapefile code was modified http://nbviewer.ipython.org/github/rjtavares/numbers_arent_people/
##blob/master/experiments/Plotting%20with%20Basemap%20and%20Shapefiles.ipynb

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import shapefile
from matplotlib.collections import LineCollection

from mpl_toolkits.basemap import Basemap
from pydap.client import open_url
from pyudl.tds import TDSCatalog

%matplotlib inline
plt.rcParams['figure.figsize'] = (20., 20.)
```

```python
##reading radar and colormap are from the ipython notebook examples from unidata at
##https://github.com/Unidata/pyudl/tree/master/examples/notebooks/radar
def radar_colormap():
    nws_reflectivity_colors = [
    "#646464", # 0
    "#04e9e7", # 5
    "#019ff4", # 10
    "#0300f4", # 15
    "#02fd02", # 20
    "#01c501", # 25
    "#008e00", # 30
    "#fdf802", # 35
    "#e5bc00", # 40
    "#fd9500", # 45
    "#fd0000", # 50
    "#d40000", # 55
    "#bc0000", # 60
    "#f800fd", # 65
    "#9854c6", # 70
    "#fdfdfd", # 75
    "#000000"
    ]

    cmap = mpl.colors.ListedColormap(nws_reflectivity_colors)

    return cmap
```

```python
##reading radar and colormap are from the ipython notebook examples from unidata at
##https://github.com/Unidata/pyudl/tree/master/examples/notebooks/radar
##get radar data from jan 27 2015
url_radar = "http://atm.ucar.edu/thredds/catalog/nexrad/composite/gini/n0r/1km/20150127/catalog.xml"
cat_radar = TDSCatalog(url_radar)
names_radar = cat_radar.datasets.keys()
names_radar.sort()

##get first image
images_radar = cat_radar.datasets[names_radar[0]]
dataset_radar = open_url(images_radar.accessUrls['OPENDAP'])

##get information from dataset
proj_attributes = dataset_radar['LambertConformal'].attributes
rsphere = proj_attributes['earth_radius']
lat_0 = proj_attributes['latitude_of_projection_origin']
lon_0 = proj_attributes['longitude_of_central_meridian']
lat_1 = proj_attributes['standard_parallel']
```

```python
##reading radar and colormap are from the ipython notebook examples from unidata at
##https://github.com/Unidata/pyudl/tree/master/examples/notebooks/radar
llcrnrlon = -120.02283
llcrnrlat = 23.0132
urcrnrlon = -63.5105
urcrnrlat = 47.30635
m = Basemap(projection='lcc', lat_0 = lat_0, lon_0 = lon_0, lat_1 = lat_1,
              llcrnrlon = llcrnrlon, llcrnrlat = llcrnrlat,
              urcrnrlat = urcrnrlat, urcrnrlon = urcrnrlon,
              area_thresh = 1000., rsphere = rsphere, resolution='i')

data_radar = np.squeeze(dataset_radar["Reflectivity"])

##get rid of noise
data_radar[data_radar<=10] = None

x_radar = dataset_radar['x'][:]
y_radar = dataset_radar['y'][:]
width = x_radar.max() - x_radar.min()
height = y_radar.max() - y_radar.min()

x_radar = (m.llcrnrx - x_radar.min()) + x_radar
y_radar = (m.llcrnry - y_radar.min()) + y_radar
```

```python
##set range of colors from 0 to 80 like a typical radar image
norm = mpl.colors.Normalize(vmin=0,vmax=80)

##read shapefile from https://mesonet.agron.iastate.edu/request/gis/watchwarn.phtml
r = shapefile.Reader(r"C:\Users\ahaberlie\.ipynb_checkpoints\wwa_201501270000")
shapes = np.array(r.shapes())
records = np.array(r.records())

##only care about shapes associated with blizzard in this case
sub_records = records[records[:,5]=='BZ']
sub_shapes = shapes[records[:,5]=='BZ']

################################################################################################
##from http://nbviewer.ipython.org/github/rjtavares/numbers_arent_people/blob/master/experiments
##/Plotting%20with%20Basemap%20and%20Shapefiles.ipynb
ax = plt.subplot(111)

for record, shape in zip(sub_records,sub_shapes):
    #read shape
    lons,lats = zip(*shape.points)
    data = np.array(m(lons, lats)).T

    #each shape may have different segments
    if len(shape.parts) == 1:
        segs = [data,]
    else:
        segs = []
        for i in range(1,len(shape.parts)):
            index = shape.parts[i-1]
            index2 = shape.parts[i]
            segs.append(data[index:index2])
        segs.append(data[index2:])

    #draw lines... make sure z order is above basemap boundaries
    lines = LineCollection(segs,antialiaseds=(1,),zorder=2)
    lines.set_facecolors('r')
    lines.set_edgecolors('k')
    lines.set_linewidth(1)
    ax.add_collection(lines)
################################################################################################

##show the radar image and put it above everything else with zorder=3
m.imshow(data_radar, norm=norm, cmap=radar_colormap(),origin='upper',zorder=3)


m.drawcoastlines(zorder=1)
m.drawstates(zorder=2)
m.drawcountries(zorder=2)
m.drawmapboundary(fill_color='gray')
m.fillcontinents(color='lightgray',lake_color='gray',zorder=1)
#plt.colorbar(ticks=np.arange(0,85,5))

plt.title("Radar Reflectivity and Blizzard Warnings - 1/27/15 00z")
```
